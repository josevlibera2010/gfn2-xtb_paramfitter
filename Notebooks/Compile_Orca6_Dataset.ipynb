## Reading all JSON files from TestData directory

```python
import json
import os
from pathlib import Path
import numpy as np
```

## Reading JSON files from TestData directory ordered by name

```python
json_files_sorted = [f'TestData/{i}.json' for i in range(1,43)]

json_data_ordered = {}
for json_file in json_files_sorted:
    with open(json_file, 'r') as f:
        json_data_ordered[json_file] = json.load(f)

print(f"Found {len(json_data_ordered)} JSON files (ordered by name):")
for filename in json_data_ordered.keys():
    print(f"  - {filename}")
```

```python
NATM=json_data_ordered['TestData/1.json']['Geometry_1']['Geometry']['NATOMS']
```

```python
energies=[(json_data_ordered[i]['Geometry_1']['SCF_Energy']['SCF_ENERGY']+json_data_ordered[i]['Geometry_1']["VdW_Correction"]["VDW"])*627.5 for i in json_data_ordered]
```

```python
charges={i.split('/')[1].replace('.json',''):np.array(json_data_ordered[i]['Geometry_1']['Mulliken_Population_Analysis']['ATOMICCHARGES']).T[0].tolist() for i in json_data_ordered}
```

```python
gradients={i.split('/')[1].replace('.json',''):np.array(json_data_ordered[i]['Geometry_1']['Nuclear_Gradient']['GRAD']).reshape((NATM,3)).tolist() for i in json_data_ordered}
```

```python
atomic_numbers={i.split('/')[1].replace('.json',''):np.array(json_data_ordered[i]['Geometry_1']['Mulliken_Population_Analysis']['ATNO']).T[0].tolist() for i in json_data_ordered}
```

```python
multiplicity={i.split('/')[1].replace('.json',''):json_data_ordered[i]['Geometry_1']["Dipole_Moment"]['MULT'] for i in json_data_ordered}
```

```python
net_charges={i.split('/')[1].replace('.json',''):json_data_ordered[i]['Geometry_1']["Dipole_Moment"]['STATE'] for i in json_data_ordered}
```

```python
coordinates={i.split('/')[1].replace('.json',''):[j[1:] for j in json_data_ordered[i]['Geometry_1']['Geometry']['Coordinates']['Cartesians']] for i in json_data_ordered}
```

```python
dataset={'system':{'chrg':net_charges,'mult':multiplicity},'energies':energies,'atomic_numbers':atomic_numbers,'gradients':gradients,'charges':charges,'coordinates':coordinates}
```

```python
with open('dataset.json', 'w') as json_file:
    json.dump(dataset, json_file, indent=4)
```
